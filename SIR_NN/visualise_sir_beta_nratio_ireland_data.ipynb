### Import Packages

```python
# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd

print("Packages Imported")
```

### Load Data

```python
# LOAD DATA
print("\nLoading Files")
data_raw = pd.read_csv('output_sir_ireland_raw.csv', delimiter=',', header=0)
data_r = data_raw.to_numpy()
print("\nTraining File Loaded")

print(data_r)
print(data_r.shape)
```

### Process Data

```python
r_days = data_r[:, 0]
r_dates = data_r[:, 1]
r_daily = data_r[:, 2]
r_active = data_r[:, 3]

print("Data Processed")
```

### Plot Daily Cases

```python
n_days = len(r_dates)
n_interval = 30

x_dates = [r_dates[x * n_interval] for x in range(0, int((n_days - 1) / n_interval) + 1)]

plt.figure(figsize=(15,8))
plt.plot(r_dates, r_daily, "-b", linewidth=3)
plt.xlabel("Date", fontsize=18)
plt.ylabel("No. Daily Cases", fontsize=18)
plt.xlim(x_dates[0], x_dates[-1])
plt.ylim(0, 10000)
plt.xticks(x_dates, rotation=50, fontsize=16, ha="right")
plt.yticks(fontsize=16)
plt.grid()
plt.show()
```

### Plot Active Cases

```python
plt.figure(figsize=(15,8))
plt.plot(r_dates, r_active, "-", linewidth=4, color="brown")
plt.xlabel("Date", fontsize=18)
plt.ylabel("No. Active Cases", fontsize=18)
plt.xlim(x_dates[0], "12/10/2021")
plt.ylim(0, 70000)
plt.xticks(x_dates, rotation=50, fontsize=16, ha="right")
plt.yticks(fontsize=16)
plt.grid()
plt.show()
```
