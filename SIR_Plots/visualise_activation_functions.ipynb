### Import Packages

```python
# Helper libraries
import numpy as np
import matplotlib.pyplot as plt

print("Packages Imported")
```

### Generate Data

```python
n_vals = 1000
n_vals_min = -5
n_vals_max = 5

x_vals = np.linspace(n_vals_min, n_vals_max, num=n_vals)

def relu(x):
    return max(0, x)

def selu(x):
    alpha = 1.6732632423543772848170429916717
    lamda = 1.0507009873554804934193349852946
    
    if (x > 0):
        y = x
    else:
        y = alpha*(np.exp(x)) - alpha
    
    return lamda * y

def tanh(x):
    return np.tanh(x)

def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

### Print Functions

```python
y_relu = [relu(x) for x in x_vals]
y_selu = [selu(x) for x in x_vals]
y_tanh = [tanh(x) for x in x_vals]
y_sigmoid = [sigmoid(x) for x in x_vals]

print("x \t\t ReLU \t\t SeLU \t\t Tanh \t\t Sigmoid")
for i in range(0, n_vals):
    print("{0:.2f} \t\t {1:.4f} \t {2:.4f} \t {3:.4f} \t {4:.4f}".format(x_vals[i], y_relu[i], y_selu[i], y_tanh[i], y_sigmoid[i]))
```

### Plot Functions

```python
xticks = np.linspace(n_vals_min, n_vals_max, num=n_vals_max - n_vals_min + 1)

fig, axs = plt.subplots(2, 2, figsize=(15,8))

axs[0, 0].plot(x_vals, y_relu, linestyle='solid', linewidth=3, color="red", label='ReLU')
axs[0, 0].set_title("ReLU", fontsize=18)
axs[0, 0].set_xlim(n_vals_min, n_vals_max)
axs[0, 0].set_ylim(-1, 5)
axs[0, 0].set_xticks(xticks)
axs[0, 0].grid()

axs[1, 0].plot(x_vals, y_tanh, linestyle='solid', linewidth=3, color="green", label='Tanh')
axs[1, 0].set_title("Tanh", fontsize=18)
axs[1, 0].set_xlim(n_vals_min, n_vals_max)
axs[1, 0].set_ylim(-1, 1)
axs[1, 0].set_xticks(xticks)
axs[1, 0].grid()

axs[0, 1].plot(x_vals, y_selu, linestyle='solid', linewidth=3, color="blue", label='SeLU')
axs[0, 1].set_title("SeLU", fontsize=18)
axs[0, 1].set_xlim(n_vals_min, n_vals_max)
axs[0, 1].set_ylim(-2, 5)
axs[0, 1].set_xticks(xticks)
axs[0, 1].grid()

axs[1, 1].plot(x_vals, y_sigmoid, linestyle='solid', linewidth=3, color="black", label='Sigmoid')
axs[1, 1].set_title("Sigmoid", fontsize=18)
axs[1, 1].set_xlim(n_vals_min, n_vals_max)
axs[1, 1].set_ylim(0, 1)
axs[1, 1].set_xticks(xticks)
axs[1, 1].grid()

for row in axs: 
    for ax in row: 
        ax.grid(b=True, which='major')

fig.tight_layout()

plt.savefig('myfigure_100.png', dpi=500)
```
