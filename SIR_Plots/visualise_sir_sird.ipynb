### Import Packages

```python
# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd

print("Packages Imported")
```

### Load Data

```python
# LOAD DATA
print("\nLoading Files")

data_sir_raw = pd.read_csv("sir_data_raw.csv", delimiter=',', header=0)
data_sir = data_sir_raw.to_numpy()
data_sird_raw = pd.read_csv("sird_data_raw.csv", delimiter=',', header=0)
data_sird = data_sird_raw.to_numpy()

print("\nTraining File Loaded")

print(data_sir)
print(data_sir.shape)

print(data_sird)
print(data_sird.shape)
```

### Process Data

```python
time = data_sir[:, 0]

sir_s = data_sir[:, 1]
sir_i = data_sir[:, 2]
sir_r = data_sir[:, 3]

sird_s = data_sird[:, 1]
sird_i = data_sird[:, 2]
sird_r = data_sird[:, 3]
sird_d = data_sird[:, 4]

N = 100000

print("Data Processed")
```

### Plot SIR

```python
plt.figure(figsize=(15,8))

plt.plot(time, sir_s, "-", color='blue', linewidth=3, label="S")
plt.plot(time, sir_i, "--", color='orange', linewidth=3, label="I")
plt.plot(time, sir_r, "-.", color='green', linewidth=3, label="R")
plt.xlabel("Day", fontsize=18)
plt.ylabel("No. Cases", fontsize=18)
plt.xlim(0, 120)
plt.ylim(0, 100000)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid()
plt.legend(loc='center right', fontsize=18)
plt.show()

print("Max Infected: \t\t {0:.0f} \t\t {1:.2f}%\n".format(max(sir_i), 100 * max(sir_i) / N))

print("Final Susceptible: \t {0:.0f} \t\t {1:.2f}%".format(sir_s[-1], 100 * sir_s[-1] / N))
print("Final Recovered: \t {0:.0f} \t\t {1:.2f}%".format(sir_r[-1], 100 * sir_r[-1] / N))
```

### Plot SIRD

```python
plt.figure(figsize=(15,8))

plt.plot(time, sird_s, "-", color='blue', linewidth=3, label="S")
plt.plot(time, sird_i, "--", color='orange', linewidth=3, label="I")
plt.plot(time, sird_r, "-.", color='green', linewidth=3, label="R")
plt.plot(time, sird_d, ".", color='red', linewidth=3, label="D")
plt.xlabel("Day", fontsize=18)
plt.ylabel("No. Cases", fontsize=18)
plt.xlim(0, 120)
plt.ylim(0, 100000)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid()
plt.legend(loc='center right', fontsize=18)
plt.show()

print("Max Infected: \t\t {0:.0f} \t\t {1:.2f}%\n".format(max(sird_i), 100 * max(sird_i) / N))

print("Final Susceptible: \t {0:.0f} \t\t {1:.2f}%".format(sird_s[-1], 100 * sird_s[-1] / N))
print("Final Recovered: \t {0:.0f} \t\t {1:.2f}%".format(sird_r[-1], 100 * sird_r[-1] / N))
print("Final Dead: \t\t {0:.0f} \t\t {1:.2f}%".format(sird_d[-1], 100 * sird_d[-1] / N))
```
