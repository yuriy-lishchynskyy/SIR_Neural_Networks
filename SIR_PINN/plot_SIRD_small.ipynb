```python
import numpy as np
import matplotlib.pyplot as plt

print("Packages Imported")
```

### Read Data

```python
data = np.genfromtxt('data_SIRD_small.csv', delimiter=',', skip_header=1)
print(data.shape)

titles = ["Beta Constant", "Beta Step x1", "Beta Step x2", "Step x1 All", "Beta Exponential Decay"]
```

### Format Data

```python
indx_real = 0 # where real data starts
indx_pred = 16 # where predicted data starts

t_real = data[:, indx_real]
t_pred = data[:, indx_pred]
```

### Plot All

```python
for i in range(0, 5):
    beta_real = data[:, indx_real + (3*i) + 1]
    gamma_real = data[:, indx_real + (3*i) + 2]
    mu_real = data[:, indx_real + (3*i) + 3]

    beta_pred = data[:, indx_pred + (3*i) + 1]
    gamma_pred = data[:, indx_pred + (3*i) + 2]
    mu_pred = data[:, indx_pred + (3*i) + 3]
    
    print(titles[i])
    
    plt.figure(figsize=(15,8)) # plot beta / gamma
    plt.plot(t_real, beta_real, "b--", label="Real Beta", linewidth=3)
    plt.plot(t_real, gamma_real, "r--", label="Real Gamma", linewidth=3)
    plt.plot(t_pred, beta_pred, "b-", label="Predicted Beta", linewidth=4)
    plt.plot(t_pred, gamma_pred, "r-", label="Predicted Gamma", linewidth=4)
    plt.xlabel("Day", fontsize=18)
    plt.ylabel("Parameter Value", fontsize=18)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.ylim(0, 0.3)
    plt.legend(loc='upper left', fontsize=16)
    plt.show()
    
    plt.figure(figsize=(15,8)) # plot mu
    plt.plot(t_real, mu_real, "g--", label="Real Mu", linewidth=3)
    plt.plot(t_pred, mu_pred, "g-", label="Predicted Mu", linewidth=4)
    plt.xlabel("Day", fontsize=18)
    plt.ylabel("Parameter Value", fontsize=18)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.ylim(0, 0.01)
    plt.legend(loc='upper left', fontsize=16)
    plt.show()
```
